```python
import xlrd

workbook = xlrd.open_workbook('Codeforces Rating 0 to 1299 Extra.xls')
worksheet = workbook.sheet_by_index(0)
print(worksheet)
```

```python

for rx in range(worksheet.nrows):
   print(worksheet.row(rx))
```

```python
file = open("myfile.md","w+")


problem_name = worksheet.col_values(0)
level = worksheet.col_values(2)
link = worksheet.col_values(3)
file.write("| {0} | {1} |\n".format(problem_name[0], level[0]))
file.write("|:---:|:---:|\n")
for i in range(1, len(problem_name)):
    file.write("| [{0}]({1}) | {2} |\n".format(problem_name[i], link[i], level[i]))
```
